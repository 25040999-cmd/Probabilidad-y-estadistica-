```python
import math as mt #Para utilizar factorial
from itertools import combinations
from itertools import combinations_with_replacement
```

```python
caracteres = ["A", "B", "C", "D", "E", "F", "0", "1", "2", "3", "4", "5" ]
n = len(caracteres)
r = 10
Cn = mt.factorial(n) / (mt.factorial(n-r) * mt.factorial(r))
print("existen", Cn , "posibles combninaciones sin repeticion del total de ", n , "caracteres ", "en grupos de ", r , "en ", r)

print("Aquí las combinaciones")
combinaciones_caracteres = list(combinations(caracteres, r))
for comnbinacion in combinaciones_caracteres:
    print(comnbinacion)
```
