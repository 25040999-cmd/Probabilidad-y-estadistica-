```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#Generar datos para la distribucion normal

media = 70  #Media
desv_std = 4  # Desviacion estandar
num_datos = 500

# generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
# # Crear un rango de valores para crear una densidad
x = np.linspace(media -3 * desv_std, media +3 * desv_std, num_datos)

    #Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi))*np.exp(-(x - media)**2 / (2 * desv_std **2))

# Para no imprimir toda la lista de valores (500), desplegamos los primeros 50

print("Imprimir los primeros 50 valores de x", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2, )

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

**PROBLEMA 3 DE TAREA**

```python
# 1. Definimos los datos del problema
N = 500           # Total de estudiantes
media = 70        # Media en kg
desviacion = 4    # Desviación estándar en kg

limite_inf = 68   # Límite inferior
limite_sup = 72   # Límite superior

# 2. Calculamos la probabilidad y la cantidad de estudiantes
probabilidad = norm.cdf(limite_sup, media, desviacion) - norm.cdf(limite_inf, media, desviacion)
cantidad_estudiantes = probabilidad * N

# 3. Imprimimos los resultados en texto
print("--- RESULTADOS ---")
print(f"Probabilidad de que un estudiante pese entre {limite_inf} y {limite_sup} kg: {probabilidad:.4f} (o {probabilidad * 100:.2f}%)")
print(f"Cantidad estimada de estudiantes: {round(cantidad_estudiantes)} estudiantes\n")
```

```python
# Generamos un rango de valores para el eje X (de 4 desviaciones estándar hacia atrás y adelante de la media)
x = np.linspace(media - 4*desviacion, media + 4*desviacion, 1000)

# Calculamos los valores de la curva (Función de Densidad de Probabilidad - pdf)
y = norm.pdf(x, media, desviacion)

# Configuramos el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Dibujamos la curva principal de la campana de Gauss
plt.plot(x, y, 'b-', linewidth=2, label=f'Distribución Normal N({media}, {desviacion})')
x_sombreado = np.linspace(limite_inf, limite_sup, 100)
y_sombreado = norm.pdf(x_sombreado, media, desviacion)
plt.fill_between(x_sombreado, y_sombreado, color='skyblue', alpha=0.6,
                 label=f'Área: {probabilidad*100:.2f}% (Aprox. {round(cantidad_estudiantes)} est.)')
plt.axvline(media, color='red', linestyle='--', linewidth=1.5, label=f'Media ($\mu$ = {media})')
plt.title('Distribución Normal de los Pesos de Estudiantes del ITD', fontsize=14, fontweight='bold')
plt.xlabel('Peso en kilogramos (kg)', fontsize=12)
plt.ylabel('Densidad de Probabilidad', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

**Problemas de Distribución Poisson y Binomial**

**Problema 1:**


La distribución de Poisson se utiliza para modelar el número de veces que ocurre un evento en un intervalo de tiempo o espacio específico, No se tiene un número fijo de ensayos "n" con probabilidades de éxito/fracaso, lo cual descartaría la distribución Binomial.


```python
import math
from scipy.stats import poisson

# Datos del problema
media_lambda = 5

print("--- SOLUCIÓN AL PROBLEMA DE DISTRIBUCIÓN DE POISSON ---\n")

# Inciso A: Probabilidad de que x = 2
x_a = 2
prob_a = poisson.pmf(x_a, media_lambda)
print("a) ¿Cuál es la probabilidad de que el número de personas sea dos?")
print(f"   La probabilidad de P(x={x_a}) es: {prob_a:.4f} o {prob_a * 100:.2f}%")
print("-" * 50)

# Inciso B: Probabilidad de que x > 10
x_b = 10
# Calculamos P(x <= 10) usando la función de distribución acumulada (cdf)
prob_acumulada_10 = poisson.cdf(x_b, media_lambda)
# P(x > 10) = 1 - P(x <= 10)
prob_b = 1 - prob_acumulada_10

print("b) ¿Es probable que x exceda de 10?")
print(f"   La probabilidad de P(x > {x_b}) es: {prob_b:.4f} o {prob_b * 100:.2f}%")

if prob_b < 0.05:
    print("   Explicación: No es probable. Como la probabilidad es menor al 5% (es del 1.37%),")
    print("   se considera un evento muy raro estadísticamente.")
else:
    print("   Explicación: Sí, existe una probabilidad considerable.")
```

a). ¿Cuál es la probabilidad de que el número de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un día particular, sea dos?

$\lambda = 5$ (media de ingresos por día)$e \approx 2.71828$$x$ es el número de eventos que queremos calcular.a). ¿Cuál es la probabilidad de que el número de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un día particular, sea dos?Aquí queremos calcular $P(x = 2)$:$$P(2) = \frac{e^{-5} \cdot 5^2}{2!} = \frac{0.0067379 \cdot 25}{2} \approx 0.0842$$

$$P(x) = \frac{e^{-\lambda} \cdot \lambda^x}{x!}$$

La probabilidad es de 0.0842, o lo que es igual, un 8.42%.

b). ¿Es probable que x exceda de 10? ExplicarAquí queremos calcular la probabilidad de que $x > 10$, lo cual matemáticamente se expresa como $P(x > 10) = 1 - P(x \le 10)$.Al sumar las probabilidades de $x=0$ hasta $x=10$ y restárselo a 1, obtenemos:$$P(x > 10) \approx 1 - 0.9863 = 0.0137$$Respuesta b): No es probable. La probabilidad de que ingresen más de 10 personas en un día es de apenas 0.0137 (aproximadamente 1.37%).

**Problema 2**

 de distribución: Binomial
 Explicación: Estrictamente, este es un experimento de tipo Binomial porque cumple con sus tres características fundamentales:Existe un número fijo de ensayos o tamaño de muestra: n = 1000 motores.Cada ensayo tiene solo dos resultados posibles: el motor es "defectuoso" (éxito para nuestro cálculo) o "no defectuoso" (fracaso).La probabilidad de que un motor sea defectuoso es constante para cada uno: p = 0.001


```python
from scipy.stats import binom

# Datos del problema
n = 1000      # Número total de motores (ensayos)
p = 0.001     # Probabilidad de que un motor sea defectuoso

print("--- SOLUCIÓN AL PROBLEMA 2 (DISTRIBUCIÓN BINOMIAL) ---\n")

# Inciso 1: Probabilidad de que ninguno (0) sea defectuoso
x_0 = 0
prob_0 = binom.pmf(x_0, n, p)
print("1. ¿Cuál es la probabilidad de que ninguno sea defectuoso?")
print(f"   P(x=0) = {prob_0:.4f}  ({prob_0 * 100:.2f}%)")

# Inciso 2: Probabilidad de que tres (3) sean defectuosos
x_3 = 3
prob_3 = binom.pmf(x_3, n, p)
print("2. ¿Cuál es la probabilidad de que tres sean defectuosos?")
print(f"   P(x=3) = {prob_3:.4f}  ({prob_3 * 100:.2f}%)")


# Inciso 3: Probabilidad de que cuatro (4) sean defectuosos
x_4 = 4
prob_4 = binom.pmf(x_4, n, p)
print("3. ¿Cuál es la probabilidad de que cuatro sean defectuosos?")
print(f"   P(x=4) = {prob_4:.4f}  ({prob_4 * 100:.2f}%)")
```

a) ¿Cuál es la probabilidad de que ninguno sea defectuoso? ($x = 0$)$$P(0) = \binom{1000}{0} (0.001)^0 (0.999)^{1000} = 1 \cdot 1 \cdot (0.999)^{1000} \approx 0.3677$$Respuesta: La probabilidad de que ninguno sea defectuoso es de 0.3677 (o 36.77%).

b) ¿Cuál es la probabilidad de que tres sean defectuosos? ($x = 3$)$$P(3) = \binom{1000}{3} (0.001)^3 (0.999)^{997} \approx 0.0613$$Respuesta: La probabilidad de que exactamente tres sean defectuosos es de 0.0613 (o 6.13%).

c) ¿Cuál es la probabilidad de que cuatro sean defectuosos? ($x = 4$)$$P(4) = \binom{1000}{4} (0.001)^4 (0.999)^{996} \approx 0.0153$$Respuesta: La probabilidad de que exactamente cuatro sean defectuosos es de 0.0153 (o 1.53%).
