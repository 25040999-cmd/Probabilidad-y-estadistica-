```python
from statistics import *
```

```python
# Lista de calificaciones del grupo a de probabilidad y estadistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
#Sacar la media de las calificaciones con una formula
promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
#Utilizar la función mean que esta en la libreria statistics para sacar el promedio
promedio2 = mean(grupoa)
print(promedio2)
```

```python
#Determinar el valor de la mediana con la funcion median

mediana = median(grupoa)
print(mediana)
```

```python
#Determinar el valor de la moda con la funcion mode
moda = mode(grupoa)
print(moda)
```

```python
#en caso de que exista mas de una moda, se puede emplear la función multimode para sacar multiples modas que puedan existir

multimoda= multimode(grupoa)
print(multimoda)
```

```python
#ejemplo 3; histograma de dos fabricas de tela
#libreria para generar numeros aleatorios
import random
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform (50,100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round(random.uniform (70,100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
#libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la Fábrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la Fábrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 =median(fabrica1)
moda1= mode(fabrica1)

print("Valores de la media de la fabrica1 es = " , media1)
print("Valores de la mediana de la fabrica1 es = ", mediana1)
print ("Valores de la moda de la fabrica1 es = ", moda1)

```

```python
#generar 2 sublot de 1 renglon por 2 columnas para desplegar los 2 histogramas,
#Una al lado de la otra
fig, (ax1, ax2) = plt.subplots(nrows =1, ncols=2, figsize= (10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
