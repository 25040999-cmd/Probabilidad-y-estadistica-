### **Programa 3: Tablas de frecuencias**

```python
import random
edades = []
for i in range(100):
  #hacemos un cliclo para generar 100 numeros aleatorios que representan las edades
  #generamos un numero aleatorio con la funcion "radint" y lo guardamos en la variable
  #número_aleatorio
    numero_aleatorio = random.randint(18, 70)
    edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los números aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
#calcular la tabla de distribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')
#Imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['frec relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['frec porcentual'] = tabla_frecuencias['Frecuencia'] /100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

# Regla de sturges
######1 .- Es una regla que siruve para calcular el número de clases o intervalos idóneo en los que se debe dividir un conjunto de datos
###### 2 Establece que el número de clases es igual a uno más el logaritmo en base dos del número total de datos
######  k=1+3,322*log10(N)




```python
#importamos la libreria match de funciones matemáticas
import math
```

```python
# Volvemos a usar la lista edades
print(edades)
```

```python
#Crear un DataFrame ocn las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
#Calcular el número de intervalos utilizando la regla de Sturges
n = len(edades)
K = int(1 + math.log2(n))
print("Valor de n = ", n , " Número de intervalos =", K)
```

```python
#Crear la tabla de frecuencia
tabla_frecuencias = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
#generar columnas siguientes
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec'] / n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Porc'].cumsum()
print(tabla_frecuencias)
```
