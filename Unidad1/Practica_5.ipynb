```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
#Cargar el archivo CSV en una variable llamada covid
covid = pd.read_csv("covid19.csv")
```

```python
#Mostramos los primeros 5 registros del archivo
covid.head()
```

```python
#Mostrar la información del archivo
#Muestra el numero de registros, tipo de datos de la columna, valores nulos, memoria etc
covid.info()
```

```python
# Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
# Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
# Dibuja el histograma del atributo "EDAD" con bins = 3
#Calcular la media y guardarla en una variable para mostrarla en el histograma
media = covid.EDAD.mean()
plt.hist(covid.EDAD, bins=11)
plt.title("Histograma de la edad pruebas COVID")
plt.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
```

```python
#dibuja el histograma de la columna sexo, esto permite coparar cuantos hay de cada grupo
plt.hist(covid.SEXO, bins = 3)
plt.title("Histograma del sexo de las pruebas COVID")
plt.xlabel('Sexo')
plt.ylabel('Frecuencia')
```

```python
#Dibuja el histograma del atributo "tipo paciente, esto permite comparar cuantos hay de cada grupo"
plt.hist(covid.TIPO_PACIENTE, bins = 3)
plt.title("Histograma del tipo de paciente de las pruebas COVID")
plt.xlabel('Tipo de paciente')
plt.ylabel('Frecuencia')
```

```python
#Crear un diagrama de cajas para visualizar la distribución de los rangos de edad por sexo
sns.boxplot(x='SEXO', y='EDAD', data=covid)
```

```python
#Crear un diagrama de cajas para visualizar la distribución de los rangos de edad
#por tipo de paciente
sns.boxplot(x='TIPO_PACIENTE', y='EDAD', data=covid)
```
