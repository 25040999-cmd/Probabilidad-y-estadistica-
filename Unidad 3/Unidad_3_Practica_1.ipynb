```python
#VALOR ESPERADO (EJEMPLO EN PYTHON)
x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]
# inicializar una variable para acumular la suma
valor_esperado = 0
#Recorrer los valores y sus probabilidades
for i in range(len(x)):
    xi = x[i] #valor de la variable aleatori
    pi = p[i] #probabilidad de ese valor
    producto = xi * pi #multiplicar valor por su probabilidad
    valor_esperado += producto # sumar al acumulador
    print(f"x = {xi}, p = {pi}, x*p = {producto:.2f}")
# Mostrar resultados final
print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

```python
#Varianza y desviación estandar
import math
x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]
#Calcular E(X)
esperanza = 0
for i in range(len(x)):
    esperanza += x[i] * p[i]
#Calcular E(X elevado a 2)
esperanza_cuadrada = 0
for i in range(len(x)):
    esperanza_cuadrada += (x[i] ** 2) * p[i]
    #Calcular varianza y desviación estandar
varianza = esperanza_cuadrada - (esperanza ** 2)
desviacion = math.sqrt(varianza)

# Mostrar resultados
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrada:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación estándar = {desviacion:.2f}")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]

# Calcular función acumulada
F = np.cumsum(p)
# Mostrar resultados
for xi, fi in zip (x, F):
  print (f"F({xi}) = {fi:.2f}")
#graficar
plt.step(x, F, where="post")
plt.title("Función de Distribución acumulada (FDA)")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```

```python
#EJERCICIOS DE TAREA
#EJERCICIO 1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Datos del problema
X = np.array([0, 1, 2, 3, 4, 5])
P_X = np.array([0.10, 0.25, 0.30, 0.20, 0.10, 0.05])

print("--- RESULTADOS DEL EJERCICIO 1 ---\n")

# 1. Verificar que las probabilidades sumen 1
suma_probabilidades = np.sum(P_X)
print(f"1. Suma de probabilidades: {suma_probabilidades:.2f}")

# 2. Elaborar la distribución de probabilidad (Tabla)
df_distribucion = pd.DataFrame({'Número de células (X)': X, 'Probabilidad P(X)': P_X})
print("\n2. Distribución de probabilidad de X:")
print(df_distribucion.to_string(index=False))

# Gráfica de la distribución de probabilidad
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(X, P_X, color='skyblue', edgecolor='black')
plt.title('Distribución de Probabilidad de X')
plt.xlabel('Número de células anormales (X)')
plt.ylabel('Probabilidad P(X)')
plt.xticks(X)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 3. Calcular Valor Esperado, Varianza y Desviación Estándar
# Valor esperado E(X) = Sumatoria de X * P(X)
E_X = np.sum(X * P_X)

# Varianza Var(X) = E(X^2) - [E(X)]^2
E_X2 = np.sum((X**2) * P_X)
varianza = E_X2 - (E_X**2)

# Desviación estándar
desv_estandar = np.sqrt(varianza)

print(f"\n3. Cálculos estadísticos:")
print(f"   Valor Esperado E(X): {E_X:.2f} células")
print(f"   Varianza Var(X): {varianza:.4f}")
print(f"   Desviación Estándar: {desv_estandar:.4f} células")

# 4. Construir y graficar la Función de Distribución Acumulada F(x)
F_x = np.cumsum(P_X)
df_acumulada = pd.DataFrame({'X': X, 'P(X)': P_X, 'F(x) Acumulada': F_x})
print("\n4. Función de Distribución Acumulada F(x):")
print(df_acumulada.to_string(index=False))

# Gráfica de la Función de Distribución Acumulada F(x)
plt.subplot(1, 2, 2)
plt.step(X, F_x, where='post', color='salmon', linewidth=2, marker='o')
plt.title('Función de Distribución Acumulada F(x)')
plt.xlabel('Número de células anormales (x)')
plt.ylabel('Probabilidad Acumulada F(x)')
plt.xticks(X)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

 El valor esperado (o esperanza matemática) es de $2.1$ células anormales por muestra. Esto significa que si el investigador analizara una gran cantidad de muestras, en promedio encontraría alrededor de 2 células anormales por cada lámina observada.



 La desviación estándar calculada es de $1.3$ células. Dado que el rango posible observado va de 0 a 5, una desviación de 1.3 indica una variabilidad moderada-baja. La mayoría de las observaciones caerán cerca del promedio (específicamente, hay un $75\%$ de probabilidad de encontrar entre 1 y 3 células anormales, sumando sus probabilidades individuales).



 La función de distribución acumulada $F(x)$ nos permite ver que es mucho más probable encontrar un número bajo de células anormales que uno alto. Por ejemplo, al observar $F(3) = 0.85$, podemos concluir que hay un $85\%$ de probabilidad de encontrar 3 o menos células anormales en una muestra. Por el contrario, la probabilidad de encontrar una cantidad alta, como 4 o más células anormales, es bastante baja (solo el $15\%$, correspondiente al complemento $1 - 0.85$).

Para determinar si un juego de azar es "favorable", "justo" o "desfavorable", debemos comparar el valor esperado con el costo del boleto (un dato que el problema no menciona explícitamente).

Si el boleto para participar en la tómbola cuesta menos de 8.50 pesos, el juego es matemáticamente favorable para el jugador (tiene una esperanza matemática positiva y ganaría dinero a largo plazo).

Si el boleto cuesta exactamente 8.50 pesos, es un juego justo.

Si el boleto cuesta más de 8.50 pesos (por ejemplo, 10 o 20 pesos), el juego es desfavorable para el jugador, ya que en promedio pagará más de lo que recuperará en premios.

```python
#Ejercicio 3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Datos del Ejercicio 3
X = np.array([0, 1, 2, 3, 4, 5]) # Unidades vendidas
P_X = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10]) # Probabilidad

print("--- RESULTADOS DEL EJERCICIO 3 ---\n")

# 1. Verificar que las probabilidades sumen 1
suma_probabilidades = np.sum(P_X)
print(f"1. Suma de probabilidades: {suma_probabilidades:.2f}")

# 2. Elaborar la distribución de probabilidad de X
df_distribucion = pd.DataFrame({'Unidades vendidas (X)': X, 'Probabilidad P(X)': P_X})
print("\n2. Distribución de probabilidad de X:")
print(df_distribucion.to_string(index=False))

# Preparamos la figura para las gráficas
plt.figure(figsize=(14, 5))

# Gráfica de la distribución de probabilidad (Barras)
plt.subplot(1, 2, 1)
plt.bar(X, P_X, color='mediumpurple', edgecolor='black')
plt.title('Distribución de Probabilidad de X (Ventas)')
plt.xlabel('Unidades vendidas (X)')
plt.ylabel('Probabilidad P(X)')
plt.xticks(X)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 3. Calcular Valor Esperado, Varianza y Desviación Estándar
# E(X) = Sumatoria de X * P(X)
E_X = np.sum(X * P_X)

# Var(X) = E(X^2) - [E(X)]^2
E_X2 = np.sum((X**2) * P_X)
varianza = E_X2 - (E_X**2)

# Desviación estándar
desv_estandar = np.sqrt(varianza)

print(f"\n3. Cálculos estadísticos:")
print(f"   Valor Esperado E(X): {E_X:.2f} unidades")
print(f"   Varianza Var(X): {varianza:.4f}")
print(f"   Desviación Estándar: {desv_estandar:.4f} unidades")

# 4. Construir y graficar la Función de Distribución Acumulada F(x)
F_x = np.cumsum(P_X)
df_acumulada = pd.DataFrame({'Unidades (x)': X, 'P(X)': P_X, 'F(x) Acumulada': F_x})
print("\n4. Función de Distribución Acumulada F(x):")
print(df_acumulada.to_string(index=False))

# Gráfica de la Función de Distribución Acumulada F(x)
plt.subplot(1, 2, 2)
# plt.step crea la gráfica de escalón
plt.step(X, F_x, where='post', color='teal', linewidth=2, marker='o')
plt.title('Función de Distribución Acumulada F(x)')
plt.xlabel('Unidades vendidas (x)')
plt.ylabel('Probabilidad Acumulada F(x)')
plt.xticks(X)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

El valor esperado calculado es de $2.8$ unidades. Esto significa que, en promedio y observando a largo plazo, la tienda espera vender aproximadamente entre 2 y 3 unidades de este producto cada semana.

La demanda tiene una variabilidad moderada a baja. La desviación estándar es de aproximadamente $1.29$ unidades. Si observamos la distribución, vemos que el $75\%$ de las probabilidades se concentran entre vender 2, 3 o 4 unidades (sumando sus probabilidades: $0.25 + 0.30 + 0.20 = 0.75$). Al haber una concentración tan alta de probabilidad alrededor de la media (2.8) y una desviación estándar de apenas un poco más de 1 unidad, podemos concluir que la demanda es relativamente estable y predecible para la tienda, siendo muy poco común tener semanas con 0 ventas o ventas máximas de 5.
