```python
from sklearn.linear_model import LinearRegression #Modelo de regresión
import numpy as np #Librería de cálculo numérico

#Datos de entrada
periodos = [2018, 2019, 2020, 2021, 2022, 2023]
inversión = np.array([2, 3, 5, 4, 11, 5] ).reshape(-1, 1)
ganancias = [20, 25, 34, 30, 40, 31]

#Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(inversión, ganancias)

#Predecir la ganancia para el año 2024
ganancia_2024 = modelo.predict( [[8]] )
print("La ganancia esperada para el año 2024 es de:" , ganancia_2024[0])
#Coeficientes de la regresion
print("Coeficientes de la regresión lineal: ", modelo.coef_[0])
print("Intercepto de la regresión lineal: ", modelo.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
#Datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
ventas = ([20, 25, 34, 30, 40, 31])
#Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(publicidad, ventas)
#Predecir la ganancia para una inversión de 8 en publicidad
inversión_predicción = np.array([[8]])
ganancia_predicción = modelo.predict(inversión_predicción)
print("La ganancia esperada para una inversión de 8 en publicidad es de:", ganancia_predicción[0])
print("Valor de la intercepción", modelo.intercept_)
print("Valor de la pendiente", modelo.coef_)
# Crear el diagrama de dispersión con la línea de regresión
plt.figure(figsize=(6, 5))
plt.scatter(publicidad, ventas, color='blue', label='Datos de entrada')
plt.plot(publicidad, modelo.predict(publicidad), color='blue', linewidth=2, label='Linea de regresión')
plt.scatter(inversión_predicción, ganancia_predicción, color='green', marker='x', label='Predicción (8,{0:.2f})'.format(ganancia_predicción[0]))
plt.title("pronostico de ganancias a partir de la inversión de publicidad")
plt.xlabel("Inversión en publicidad")
plt.ylabel("Ganancia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

# Datos de entrada
publicidad = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

#Calcular el coeficiente de correlación de Pearson
coeficiente_correlación, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlación de Pearson:", coeficiente_correlación)
print("Valor p:", p_valor)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Lista con los nombres de los archivos exactos que pide la práctica
archivos = ['grupo1.csv', 'grupo2.csv', 'grupo3.csv']

for archivo in archivos:
    print(f"--- Analizando datos del archivo: {archivo} ---")

    # 1. Carga y Exploración de Datos
    # Aquí va a leer los archivos reales que tú subiste a la carpeta de Colab
    df = pd.read_csv(archivo)

    print("Descripción estadística inicial:")
    print(df.describe())
    print("\n")

    # Separamos las variables (X = Horas, Y = Calificación)
    X = df[['Horas de Estudio']]
    Y = df['Calificación Final']

    # Generar el histograma de calificaciones
    plt.figure(figsize=(6, 4))
    plt.hist(Y, bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Histograma de Calificaciones - {archivo}')
    plt.xlabel('Calificación Final')
    plt.ylabel('Frecuencia')
    plt.show()

    # 2. Modelo de Regresión Lineal
    modelo = LinearRegression()
    modelo.fit(X, Y)

    # Obtener pendiente e intercepto
    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    print(f"Ecuación de la línea de regresión: Y = {pendiente:.4f}X + {intercepto:.4f}")

    # 3. Cálculo de la Correlación de Pearson
    correlacion = df['Horas de Estudio'].corr(df['Calificación Final'])
    print(f"Coeficiente de correlación de Pearson: {correlacion:.4f}\n")

    # 4. Visualización de Resultados
    plt.figure(figsize=(6, 4))

    # Gráfico de dispersión (scatter plot)
    plt.scatter(X, Y, color='blue', label='Datos reales del grupo')

    # Línea de regresión
    Y_pred = modelo.predict(X)
    plt.plot(X, Y_pred, color='red', linewidth=2, label='Línea de regresión')

    plt.title(f'Relación entre Horas de Estudio y Calificaciones ({archivo})')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificación Final')
    plt.legend()
    plt.show()
    print("\n")
```
